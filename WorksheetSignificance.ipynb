** Hypothesis Testing

```python
#test whether the mean murder rate is 8.5
murder_data=state_pd['Murder'].values
murder_data
```

```python
np.mean(murder_data)
```

```python
t_stat=(np.mean(murder_data)-8.5) / (np.std(murder_data,ddof=1) / np.sqrt(len(murder_data)))
```

```python
df=len(murder_data)
```

```python
p_value=2*stats.t.sf(np.abs(t_stat),df)
```

```python
ci=stats.t.interval(0.95,df,
                    loc=np.mean(murder_data),
                    scale=np.std(murder_data,ddof=1)/np.sqrt(len(murder_data))
)
```

```python
print("Mean:",np.mean(murder_data))
print("Sample Standard Deviation:",np.std(murder_data))
```

```python
print("Test Statistics:",t_stat)
```

t_stat = 2.149 This means : The Sample mean -2.149 standard errors away from 8.5 Negative sign means the sample mean is Below 8.5 The larger the |t|, the more unusual our sample is

|t| = 0.5 ---> Very close to hypothesized mean (not unusual) |t| = 1.5 ---> Moderately far from hypothesized mean |t| = 2.1 ---> Far from hypothesized np.mean |t| = 5.0 ---> Very far from hypothesized mean


```python
print("Confidence Interval",ci)
```

```python
print("p-value",p_value)
```

```python
print(f"we {'REJECT HO' if p_value <= 0.05 else 'FAIL TO REJECT HO'}")
```
